## US Treasury rates

Here we use the Quandl package to get US Treasury data and plot a yield curve. 

For the Black Scholes problem we will just take today's 1 year rate.


```python
#Importing US Treasury rates and choosing 1 year return for use in BS

import quandl as qd #Quandl for downloading financial data
import pandas as pd #Pandas for access to dataframes
import numpy as np #Numpy for access to arrays and vectorization of calculaitons
yield_=qd.get("USTREASURY/YIELD")
yield_
```

```python
today = yield_.iloc[-1,:]
today.plot(title='Treasury Yield Curve, %')
today
```

```python
rate=today["1 YR"]
print(rate)
```
