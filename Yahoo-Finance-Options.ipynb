## Yahoo-Finance-Options

This notebook downloads option data from Yahoo Finance using the yahoo_fin package.

We pick out a single case to compare the actual price on the market against the price predicted by Black-Scholes using volatility derived from returns on the underlying stock and a rate taken from the US Treasury.


```python
#http://theautomatic.net/2019/04/17/how-to-get-options-data-with-python/
#https://algotrading101.com/learn/yahoo-finance-api-guide/
from yahoo_fin import options
import pandas as pd
```

```python
ticker="nflx" # Netflix: change this to try out any other stocks available on Yahoo Finance

call_dates= options.get_expiration_dates(ticker)
call_dates
```

```python
import datetime
today = datetime.date.today()
#maturity = datetime.date(2020, 10, 23)    # Y M D  ******* NB: PICK DATE FROM ABOVE*****
maturity = datetime.date(2022, 9, 16)    # Y M D
daycount = (maturity - today).days
print('maturity ',maturity,'    daycount ',daycount)


calls=options.get_calls("nflx", maturity)
calls["Volume"]=pd.to_numeric(calls["Volume"],errors='coerce') # coerce replaces - with NaN
calls[calls.Volume>10]
```

```python
mystrike=570
mycall=calls[calls.Strike == mystrike]
mycall
```

```python
#myprice=float(0.5*(mycall["Bid"]+mycall["Ask"]))
myprice=0.5*(float(mycall.iloc[0]['Bid'])+float(mycall.iloc[0]['Ask']))
print("Maturity in",daycount,"days", "Strike", mystrike, "Price",myprice)
```
